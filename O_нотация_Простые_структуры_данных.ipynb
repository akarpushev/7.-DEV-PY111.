### Понятие алгоритма

**Алгоритм** – набор инструкций, описывающий порядок действий  
для достижения результата.

При наличии двух алгоритмов, выдающих при  
**одинаковых входных данных одинаковые результаты**,  
обычно предпочитают тот, который **быстрее**.

---

***Что такое быстрее??? Как считать??? Что такое эффективнее???***  
Алгоритмы сравнивают по двум параметрам:
1. Пространственным затратам
1. Временн**ы**м затратам

````python
list_numbers = [1, 2, 3, 4, 5]

new_list = [num ** 2 for num in list_numbers]  # сделали вторую последовательность
print(sum(new_list))
````

````python
list_numbers = [1, 2, 3, 4, 5]

sum_ = 0
for num in list_numbers:
    sum_ += num ** 2
print(sum_)
````

````python
list_numbers = [1, 2, 3, 4, 5]

new_list = (num ** 2 for num in list_numbers)  # использовали выражение-генератор
print(sum(new_list))
````

````python
%%timeit
sum([num ** 2 for num in range(6)])
````

````python
%%timeit
sum_ = 0
for num in range(6):
    sum_ += num ** 2
````

````python
%%timeit
sum(num ** 2 for num in range(6))
````

### О-большое, омега-большое и тета

- Ο (O-большое) описывает **верхнюю границу сложности**  
или рассматривает **худший случай**
- Ω (Омега-большое) - описывает **нижнюю границу сложности**  
или рассматривает **лучший случай**
- Θ (Тета) - описывает **точную оценку сложности**  
или рассматривает **средний случай**

В курсе будем говорить про **O-большое**.  
О-большое позволяет оценить как  
простаранственные так и временные затраты.

Чаще всего О-большое описывает временные затраты.  
А именно оценивает насколько быстро работает тот или иной алгоритм.  
**При этом О-большое оценивает не время,  
а как зависит количество необходимых операций  
от размера входных данных.**

А точнее как быстро растёт количество операций  
в зависимости от количества данных.


Рассмотрим такой пример:

Найти длину нумерованного.  
1. ...
1. ...
1. ...  
1. ...
1. ...

Смотрим на последний номер и отвечаем.  
Считать не надо, поэтому временные **затраты константные**,  
т.е. не зависят (почти) от размера входных данных.  
Условно говоря, какой длины не был бы список,  
просто отмотай вниз и назови размер.

---

Найти длину маркированного списка.  
* ...
* ...
* ...
* ...
* ...

Чтобы найти длину списка, надо считать каждую строчку.  
То есть нужно выполнить N операций, где N размер списка.  
Чем больше список, тем больше будет операций.  
При этом количество операций зависит от  
размера входных данных **линейно**.

---

Найти длину маркированного списка с вложенными пунктами.  
* ...
    * ...
    * ...
* ...
    * ...
    * ...

Чтобы найти длину списка, надо считать каждую строчку  
и дополнительно количество подпунктов.  
То есть нужно выполнить N * M операций,  
где N количество пунктов, M - количество подпунктов в каждом пункте.  

Чем больше список и подпунктов, тем больше будет операций.  
При этом количество операций зависит от  
размера входных данных уже **квадратично**.

1. $O(1)$ – константантная сложность
1. $O(log(N))$ - логарифмическая сложность
1. $O(N)$ - линейная сложность
1. $O(N*log(N))$  
1. $O(N^k)$, k > 1 - степенная (полиномиальная) сложность  
1. $O(k^N)$, k > 1 - показательная (экпоненциальная) сложность  
1. $O(N!)$

image.png

1. $O(1)$ - Константная сложность  
    В Python получение длины от списка
1. $O(N)$ - Линейная сложность  
    Линейные поиск.  
    В Python оператор in для проверки вхождения элемента в список.

1. $O(log_k(N)), k > 1$ - логарифмическая сложность  
    Если количество обрабатываемых элементов  
    на каждом шаге уменьшается в k раз,  
    то сложность логарифмическая $O(log_k(N))$

    Пример, бинарный поиск в отсортированных последовательностях.
    $O(log_2(N)) \to O(log(N))$

1. $O(N * log N)$  
    Эффективные алгоритмы сортировки.  
    Например, быстрая сортировка, сортировка слиянием.

1. $O(N ^ k)$  - Полиномиальная сложность  
    - $O(N ^ 2)$ - квадратичная сложность  
    - $O(N ^ 3)$ - кубическая сложность

    Нэффективные алгоритмы сортировки.  
    Например, сортировка пузырьком.

1. $O(k ^ N), k > 1$ - Экпоненциальная сложность.  
    Если количество обрабатываемых элементов на каждом шаге увеличивается в N раз,  
    то сложность показательная $O(2^N)$.  
    Чаще относится к рекурсивным функциям вычисляется сложнее  
    (рассмотрим рекурсии позже).

### Правила вычисления

1. Множители представленные в виде констант можно отбрасывать  
    $O(5) \to O(5 * 1) \to O(1)$  
    $O(100N) \to O(N)$  
    $O(100500N^2) \to O(N^2)$

    Но (!),  
    $O(N * logN)$, т.к. не константа

1. Среди двух слагаемых выбирают доминирующий оператор (см. таблицу)  
    $O(1) + O(N) \to O(N)$  
    $O(N) + O(N^2) \to O(N^2)$  
    $O(100500 * N) + O(N^2) \to O(N^2)$

1. Сложность алгоритма внутри цикла умножается на сложность цикла
```python
for i in range(n):  # O(N)
    for j in range(n):  # O(N)
        print(i + j)
```
$O(N) * O(N) \to O(N^2)$

1. Сложность последовательных алгоритмов складывается
```python
for i in range(n):  # O(N)
    ...
for j in range(n):  # O(N)
    ...
```
$O(N) + O(N) \to O(2N) \to O(N)$

1. Если на каждом шаге размер входных данных уменьшается в k раз,  
то сложность логарифмическая
```python
i = 16
while i > 1:
    print(i)
    i /= 2  # 16, 8, 4, 2, 1
```

1. Если в явном виде не указано отношение двух величин, то лучше оставить обе.  
Найти длину маркированного списка с вложенными пунктами.  
* ...
    * ...
    * ...
* ...
    * ...
    * ...

$O(N * M)$, где N - количество пунктов,  
M - максимальное количество подпунктов среди всех пунктов.

Но если скажут, что количество подпунктов  
пренебрежимо мало по отношению к количеству пунктов,  
то M можно отбросить.  
$O(N * M) \to O(N)$, т.к. M пренебрежимо мало по отношению к N

> Согласно одной из легенд, шахматы придумал мудрец по имени Сисса,  
который показал своё изобретение правителю страны.  
Тому так понравилась игра, что он дал изобретателю право самому

##### Решение

````python
count_grains = 0
for ceil in range(64):
    count_grains += 2 ** ceil

print(count_grains)
````

Массив представляет из себя N-мерную плоскую таблицу в памяти

#### Доступ к элементу

Операция доступа имеет сложность $O(1)$

`array[4]` эквивалентно инструкции

Поэтому в массивах нумерация с 0 :)

#### Вставка и удаление

При вставке происходят следующие операции:
1. Выделить дополнительную память после массива равную  
«размер ячейки * количество вставляемых элементов»
2. Все элементы, начиная с места вставки,  
    **сместить вправо** на «количество вставляемых элементов»
3. На освободившееся место вставить элемент(ы)

### Массив VS Список

|                 Массив                |                Список                |
|:-------------------------------------:|:------------------------------------:|
|           Однородные данные           |          Разнородные данные          |
|   Быстрый доступ (и замена) **O(1)**  | Медленный доступ (и замена) **O(N)** |
| Медленная вставка и удаление **O(N)** |  Быстрая вставка и удаление **O(1)** |

Если обратиться к официальной [документации](https://wiki.python.org/moin/TimeComplexity/) python по **python list**, то  
- Вставка O(N)
- Удаление O(N)
- Доступ O(1)
- Замена O(1)

Стек - абстрактный тип данных

(может в памяти и по факту быть реализован по разному,  
как массивом, так и списком, так и чем то еще).

Стек работает по принципу `LIFO – Last In First Out`

Реализует следущие операции:
- `push` - положить значение на вершину стека
- `pop` - снять значение с вершины стека
- иногда еще `peek` - подсмотреть значение внутри стека без изъятия элемента.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/b/b4/Lifo_stack.png/700px-Lifo_stack.png)

Где применяется?
- вызовы функций

- помощь в реализации других алгоритмов (BFS, DFS)

### Очередь (Queue)

Тоже абстрактный тип данных.  
Может быть реализована через массив, список, ...  
Очередь организована по принципу **`FIFO – First In First Out`**

![](https://dev.koshovyi.com/wp-content/uploads/2020/09/Fifo_queue.jpg)

Реализует две операции:
- `Enqueue` – добавить в конец очереди
- `Dequeue` – выдать из начала очереди

Вопросы:

1. Стоимость операции `enqueue`?

2. Стоимость операции `dequeue`?


#### Очередь с приоритетами

У каждого элемента в очереди есть приоритет,  
который указывается при добавлении (enqueue).

Операция dequeue выполняет возврат элемента с  
максимально возможным приоритетом, добавленным первым.  
Все еще FIFO.

Реализация: очередь одна, элементы упорядочены по приоритету.  
Нужно быстро получить, но есть время для добавления.

![](https://docs.microsoft.com/ru-ru/azure/architecture/patterns/_images/priority-queue-pattern.png)

Вопросы:

1. Стоимость операции `enqueue`?

2. Стоимость операции `dequeue`?




Реализация: количество очередей равно количеству приоритетов.  
Нужно быстро поместить элемент в очередь, но есть время для изъятия.

![](https://docs.microsoft.com/ru-ru/azure/architecture/patterns/_images/priority-queue-separate.png)
